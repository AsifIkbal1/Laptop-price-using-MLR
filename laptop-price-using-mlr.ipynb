```python
df.drop(['laptop_ID'], axis=1, inplace=True)
df.head()
```

```python
df['Ram'] = df['Ram'].str.replace('GB','')
df['Weight'] = df['Weight'].str.replace('kg','')
df['Ram']=df['Ram'].astype(int)
df['Ram']=df['Ram'].astype(float)
```

```python
def convert_memory(x):
    tmp=x.split(' ')[0]
    specs=x.split(' ')[0][-2:]
    mem=float(x.split(' ')[0][:-2])
    if(specs=='TB'):
        mem*=1000
    return float(mem)
df['Memory_Type']=df['Memory'].apply(lambda x: x.split(' ',1)[1:][0])
df['Memory_gb']=df['Memory'].apply(convert_memory)
df.drop(columns=['Memory'],inplace=True)
```

```python
categorical_features =df.columns[(df.dtypes == 'object') == True].to_list()
print(categorical_features)
```

```python
#lable encoding
for feature in categorical_features:
    uniq = df[feature].unique()
    
    new_feature = []
    for el in df[feature]:
        new_feature.append(len(uniq) - np.where(uniq==el)[0][0])
    
    df[feature] = new_feature
    
df.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop(['Price_euros'], axis=1), df['Price_euros'], test_size=0.3, random_state=0)
Model = LinearRegression().fit(X_train, Y_train)
```

```python
print('r2 score: ', metrics.r2_score(Y_test, Model.predict(X_test)))
```

```python
plt.scatter(np.linspace(0, len(Model.predict(X_test)), len(Model.predict(X_test))), Model.predict(X_test), alpha=0.5, label='predictions',color='red',s=50)
plt.scatter(np.linspace(0, len(Y_test), len(Y_test)), Y_test, alpha=0.5, label='true',color='blue',marker="+",)
plt.legend()
```
